```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values of n
n_values = [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]

# Loop through the values of n
for n in n_values:
    # Simulate throwing two dice n times
    dice_sum = np.random.randint(1, 7, size=n) + np.random.randint(1, 7, size=n)

    # Compute the frequencies of the dice sums
    hist, bins = np.histogram(dice_sum, bins=np.arange(2, 14))

    # Plot the histogram
    plt.bar(bins[:-1], hist / n)

    # Set the title of the plot
    plt.title(f"Histogram of Dice Sums (n = {n})")

    # Show the plot
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv('weight-height.csv')

plt.figure(figsize=(8, 6))
plt.scatter(data['Height'], data['Weight'], alpha=0.5)
plt.title('Height vs Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.grid(True)
plt.show()

X = data[['Height']]
y = data['Weight']

model = LinearRegression()
model.fit(X, y)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.5)
plt.plot(X, model.predict(X), color='red', linewidth=2)
plt.title('Linear Regression: Height vs Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.grid(True)
plt.show()

y_pred = model.predict(X)

rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print("RMSE:", rmse)
print("R2 Score:",r2)
```
